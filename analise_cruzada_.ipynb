```python
#Parte 1: Dados
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Dados para Pessoa A (Marketing)
dados_marketing = pd.DataFrame({
'investimento_ads_dia': [10, 15, 20, 22, 25, 30, 33, 35, 40, 41, 45, 50, 55, 60, 62],
'visitantes_unicos': [1100, 1200, 1500, 1600, 1550, 1800, 1900, 2000, 2200, 2150, 2400, 2800, 3100, 3500, 3400]
})
# Dados para Pessoa B (Logística)
dados_logistica = pd.DataFrame({
'transportadora_X': [3, 4, 4, 5, 5, 5, 5, 6, 6, 6, 7, 7, 8, 8, 9],
'transportadora_Y': [1, 2, 2, 3, 4, 5, 5, 6, 7, 8, 9, 10, 10, 11, 12]
})
```

# Parte 2: Análise Individual
Cada analista foca no seu problema:

## Pessoa A (Marketing):

 Dataset: dados_marketing

 Estatística: Calcule média e mediana das duas colunas.

 Visualização: Crie um sns.regplot para analisar a relação entre investimento_ads_dia e

visitantes_unicos.

 Conclusão: Existe correlação? A regressão linear é um bom modelo para prever visitas?

```python
#calculos
media_investimento = dados_marketing['investimento_ads_dia'].mean()
mediana_investimento = dados_marketing['investimento_ads_dia'].median()
media_visitantes = dados_marketing['visitantes_unicos'].mean()
mediana_visitantes = dados_marketing['visitantes_unicos'].median()
#visualização
sns_regplot_logistica = sns.regplot(x='investimento_ads_dia', y='visitantes_unicos', data=dados_marketing)
plt.show()
print('Média de investimento:', media_investimento)
print('Mediana de investimento:', mediana_investimento)
print(f'Média de visitantes: {media_visitantes:.2f}')
print('Mediana de visitantes:', mediana_visitantes)
```

Resposta:
As colunas de investimento e visitantes estão diretamente relacionadas o que quer dizer que se aumentarmos uma a outra terá mais chances de aumentar também.


## Pessoa B (Logística):

 Dataset: dados_logistica

 Estatística: Calcule média e desvio padrão das duas colunas.

 Visualização: Crie um sns.boxplot para comparar a distribuição das transportadoras.

 Conclusão: Qual transportadora é melhor? Justifique usando a média (velocidade) e o desvio padrão

(consistência).

```python
#calculos
media_transportadora_X = dados_logistica['transportadora_X'].mean()
media_transportadora_Y = dados_logistica['transportadora_Y'].mean()
desvio_transportadora_X = dados_logistica['transportadora_X'].std()
desvio_transportadora_Y = dados_logistica['transportadora_Y'].std()
#visualização
sns_boxplot_marketing = sns.boxplot(data=[dados_logistica['transportadora_X'], dados_logistica['transportadora_Y']])
plt.show()
print(f'Média da transportadora X: {media_transportadora_X:.2f}')
print(f'Média da transportadora Y: {media_transportadora_Y:.2f}')
print(f'Desvio padrão da transportadora X: {desvio_transportadora_X:.2f}')
print(f'Desvio padrão da transportadora Y: {desvio_transportadora_Y:.2f}')
```

Resposta:
Mesmo que a transportadora Y seja mais rapida a transportadora X tem quase a mesma velocidade de entrega e ainda tem uma consistencia significativamente melhor.
