```python
#1  Importar bibliotecas.



#2  Imagine que voc recebeu um pequeno conjunto de dados sobre imveis em uma determinada regio. O conjunto de dados contm o tamanho do imvel em metros quadrados e o seu preo de venda.



#3  Criar Histograma e Grfico de Disperso.



#1  Importar bibliotecas.

import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns



#2  Imagine que voc recebeu um pequeno conjunto de dados sobre imveis em uma determinada

#regio. O conjunto de dados contm o tamanho do imvel em metros quadrados e o seu preo de

#venda.

dados_imoveis = pd.DataFrame({'tamanho_metros_quadrados': [100,180,68,53,120,],

'preo_venda': [500.000,800.000,380.00,200.000,300.000]})



#3  Criar Histograma e Grfico de Disperso.

# O que mostra:
#Mostra como os valores de uma varivel se distribuem em diferentes faixas (bins).
#Indica quantas vezes cada intervalo aparece, ajudando a ver onde os dados se concentram e como
#variam.
# Pergunta que responde:
#Onde esto concentrados os valores?
#Exemplo: A maioria dos salrios est nas faixas baixas, mdias ou altas?

# Criando um histograma para a varivel salrio

plt.figure(figsize=(10, 6)) # Define o tamanho da figura

sns.histplot(dados_imoveis['preo_venda'], kde=True) # kde=True desenha uma linha de densidade

plt.title('Imoveis metros quadrados x valor.')

plt.xlabel('tamanho_metros_quadrados')

plt.ylabel('preo_venda')

plt.show() # Exibe o grfico

#4. Grfico de Disperso (Scatter Plot): Investigando Relaes
# O que mostra:
#Exibe a relao entre duas variveis numricas.
#Cada ponto representa um registro da base de dados  um valor no eixo X e outro no eixo Y.
#Ajuda a enxergar padres, tendncias ou relaes (positivas, negativas ou inexistentes) entre as
#variveis.
# Pergunta que responde:
#As duas variveis esto relacionadas?
#Exemplo: Quando os anos de experincia aumentam, o salrio tambm cresce?

# Criando um grfico de disperso

plt.figure(figsize=(10, 6))

sns.scatterplot(x='tamanho_metros_quadrados', y='preo_venda', data=dados_imoveis, s=100)

plt.title('Relao entre Metros quadrados e valor dos imoveis.')

plt.xlabel('tamanho_metros_quadrados')

plt.ylabel('preo_venda')

plt.grid(True)

plt.show()
```
