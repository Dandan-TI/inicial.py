```python
#1 – Importar bibliotecas.



#2 – Imagine que você recebeu um pequeno conjunto de dados sobre imóveis em uma determinada região. O conjunto de dados contém o tamanho do imóvel em metros quadrados e o seu preço de venda.



#3 – Criar Histograma e Gráfico de Dispersão.



#1 – Importar bibliotecas.

import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns



#2 – Imagine que você recebeu um pequeno conjunto de dados sobre imóveis em uma determinada

#região. O conjunto de dados contém o tamanho do imóvel em metros quadrados e o seu preço de

#venda.

dados_imoveis = pd.DataFrame({'tamanho_metros_quadrados': [100,180,68,53,120,],

'preço_venda': [500.000,800.000,380.00,200.000,300.000]})



#3 – Criar Histograma e Gráfico de Dispersão.

# O que mostra:
#Mostra como os valores de uma variável se distribuem em diferentes faixas (bins).
#Indica quantas vezes cada intervalo aparece, ajudando a ver onde os dados se concentram e como
#variam.
# Pergunta que responde:
#“Onde estão concentrados os valores?”
#Exemplo: “A maioria dos salários está nas faixas baixas, médias ou altas?”

# Criando um histograma para a variável salário

plt.figure(figsize=(10, 6)) # Define o tamanho da figura

sns.histplot(dados_imoveis['preço_venda'], kde=True) # kde=True desenha uma linha de densidade

plt.title('Imoveis metros quadrados² x valor.')

plt.xlabel('tamanho_metros_quadrados')

plt.ylabel('preço_venda')

plt.show() # Exibe o gráfico

#4. Gráfico de Dispersão (Scatter Plot): Investigando Relações
# O que mostra:
#Exibe a relação entre duas variáveis numéricas.
#Cada ponto representa um registro da base de dados — um valor no eixo X e outro no eixo Y.
#Ajuda a enxergar padrões, tendências ou relações (positivas, negativas ou inexistentes) entre as
#variáveis.
# Pergunta que responde:
#“As duas variáveis estão relacionadas?”
#Exemplo: “Quando os anos de experiência aumentam, o salário também cresce?”

# Criando um gráfico de dispersão

plt.figure(figsize=(10, 6))

sns.scatterplot(x='tamanho_metros_quadrados', y='preço_venda', data=dados_imoveis, s=100)

plt.title('Relação entre Metros quadrados e valor dos imoveis.')

plt.xlabel('tamanho_metros_quadrados')

plt.ylabel('preço_venda')

plt.grid(True)

plt.show()
```
